```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import word2number
```

```python
req = pd.read_csv("/content/hiring.csv")
req
```

```python
import math
median_testscore = math.floor(req.interview_score.median())
median_testscore
```

```python
req.test_score = req.test_score.fillna(median_testscore)
req
```

```python
req.experience
```

```python
from word2number import w2n
```

```python
req.experience = req.experience.fillna("zero")
req
```

```python
req.experience = req.experience.apply(w2n.word_to_num)
req
```

```python
res = linear_model.LinearRegression()
res.fit(req[['experience','test_score','interview_score']],req.salary)
```

```python
res.coef_
```

```python
res.intercept_
```

```python
res.predict([[2,9,6]])
```

```python
res.predict([[12,10,10]])
```
